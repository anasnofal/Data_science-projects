## Modeling

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error

def evaluate_preds(y_true, y_pred,name):

  # Calculate various evaluation metrics
  mae = mean_absolute_error(y_true, y_pred)
  R2=r2_score(y_true, y_pred)
  mse = mean_squared_error(y_true, y_pred)
  rmse = mse**0.5
  mape = mean_absolute_percentage_error(y_true, y_pred)


  metrics = {'MAE': mae,
             'MSE': mse,
             'RMSE': rmse,
             'MAPE': mape,
            'R2': R2}
  metrics['model']=name
  return metrics


def display_scores(y_test, pred):
  plt.scatter(y_test, pred,alpha=0.9)
  plt.xlabel("Actual")
  plt.ylabel("Predicted")
  x_lim = plt.xlim()
  y_lim = plt.ylim()
  plt.plot(x_lim, y_lim, "k--")
  plt.show()
  test_model_mse = mean_squared_error(y_test, pred)
  print('********************************************')
  print('the r2 value is',r2_score(y_test, pred))
  print('the rmse value is',test_model_mse**0.5)
  print('the mae value is',mean_absolute_error(y_test, pred))



def plot_sampile(y_test, pred,format='.-'):
  y_test_t=y_test.reset_index()
  arr=np.array(range(len(y_test_t)))
  plt.plot(arr,y_test, '.-', label='measured')
  plt.plot(arr,pred, 'x-', label='predicted')
  plt.xlabel('sampile')
  plt.ylabel('FS')
  plt.legend(fontsize=14)
  plt.grid(True)
```

#### random forrest 

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 500, num = 50)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()

rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train, y_train)
rf_random.best_params_
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,AlphaDropout
```

```python
model = Sequential()
model.add(Dense(units=7,activation='selu'))
model.add(AlphaDropout(rate=0.2))
model.add(Dense(units=20,activation='selu'))
model.add(AlphaDropout(rate=0.2))
model.add(Dense(units=10,activation='selu'))
model.add(AlphaDropout(rate=0.2))
model.add(Dense(units=5,activation='selu'))
model.add(AlphaDropout(rate=0.2))
model.add(Dense(units=1))
```

```python
model.compile(optimizer='adam',
    loss='mse',
    metrics=[tf.keras.metrics.MeanSquaredError()])



checkpoint_filepath = 'model.h5'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath,
    save_weights_only=True,
    monitor='val_mean_squared_error',
    mode='min',
    save_best_only=True)
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=1000,batch_size=108,callbacks=[model_checkpoint_callback])
```

```python
model.load_weights(checkpoint_filepath)
```

```python
pred_SVR_test=model.predict(X_test)
pred_SVR_X_train=model.predict(X_train)
SVR_res=evaluate_preds(y_test, pred_SVR_test,'DL')
testing = testing._append(pd.Series(SVR_res),ignore_index=True )
testing
```
