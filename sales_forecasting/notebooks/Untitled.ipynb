```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns



import pickle
```

```python
data = pd.read_csv('NLTP_F_DAILY_Makkah.csv')
data['date']=pd.to_datetime(data['date'])
df=data[['Date','System_City_Calculated']]
df.index=df['Date']
df=df.resample('1M').mean()
```

```python
df['Date']=df.index
df
```

```python
def generate_supervised(data):
    supervised_df = data.copy()
    
    #create column for each lag
    for i in range(1,13):
        col_name = 'lag_' + str(i)
        supervised_df[col_name] = supervised_df['System_City_Calculated'].shift(i)
    
    #drop null values
    supervised_df = supervised_df.dropna()
    
    return supervised_df
```

```python
generate_supervised(df)
```
